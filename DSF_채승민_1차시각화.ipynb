# 5등급제에서 장애인들이 구별로 얼마나 있는가 for 15등급과 매칭 벗 실패

```python
rank = pd.read_csv('서울특별시_ 장애유형별, 등급별 장애인등록현황_내가 수정한거.csv', encoding = 'cp949')
rank
```

```python
sum1 = 0
sum2 = 0
sum3 = 0
sum4 = 0
sum5 = 0
sum6 = 0

for i in range(0,24):
  for j in range(0,15):
    if "1" in rank.columns[j]:
      sum1 = sum1 + rank.iloc[i,j]
    elif "2" in rank.columns[j]:
      sum2 = sum2+ rank.iloc[i,j]
    elif "3" in rank.columns[j]:
      sum3 = sum3+ rank.iloc[i,j]
    elif "4" in rank.columns[j]:
      sum4 = sum4+ rank.iloc[i,j]
    elif "5" in rank.columns[j]:
      sum5 = sum5+ rank.iloc[i,j]
    elif "6" in rank.columns[j]:
      sum6 = sum6+ rank.iloc[i,j]
  j=0

print(sum1, sum2, sum3, sum4, sum5, sum6)
#123456등급의 % 계산
print(sum1/(sum1+sum2+sum3+sum4+sum5+sum6))
```

```python
data = [sum1, sum2, sum3, sum4, sum5, sum6]
labels = ["1등급","2등급","3등급","4등급","5등급","6등급"]
plt.pie(data, labels=labels, autopct='%.1f%%')
plt.show()
```
